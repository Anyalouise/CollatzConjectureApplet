```python
import matplotlib.pyplot as plt
import numpy as np
import math
import random
import sys
```

```python
sequence = []
def run(start_num):
    max_num = start_num
    curr = start_num
    sequence.append(curr)
    
    while curr != 1:
        if curr%2 == 0:
            curr = curr/2
            sequence.append(curr)
        else:
            curr = (3*curr)+1
            sequence.append(curr)
            if curr > max_num:
                max_num = curr
    return max_num
```

```python
rand_seed = random.randint(1, sys.maxsize)
print("Seed is", rand_seed)
max_num = run(rand_seed)

plt.plot(np.arange(0, len(sequence)), sequence)
plt.title('Collatz Conjecture')
plt.xlabel('Step')
plt.ylabel('Current Value')
plt.show()

s = [math.log10(i) for i in sequence]
plt.plot(np.arange(0, len(sequence)), s)
plt.title('Collatz Conjecture (log)')
plt.xlabel('Step')
plt.ylabel('Current Value')
plt.show()
sequence = []
```
